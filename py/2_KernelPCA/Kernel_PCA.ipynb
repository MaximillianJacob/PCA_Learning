```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA, KernelPCA
from sklearn.datasets import make_circles
```

```python
#Setting the seed for replicability
np.random.seed(7)
```

```python
#Generating the samples
X, y = make_circles(n_samples=500, factor=0.2, noise=0.04)
```

```python
#Performing PCA transformation
pca = PCA()
X_pca = pca.fit_transform(X)
```

```python
#Performing Kernel PCA
kernel_pca = KernelPCA(kernel='rbf', fit_inverse_transform=True, gamma=10)
X_kernel_pca = kernel_pca.fit_transform(X)
X_inverse=kernel_pca.inverse_transform(X_kernel_pca)
```

```python
#Plotting the original data
class_0 = np.where(y==0)
class_1 = np.where(y==1)
plt.figure()
plt.title("Original data")
plt.plot(X[class_0, 0], X[class_0, 1], 'bo', mfc='none')
plt.plot(X[class_1, 0], X[class_1, 1], 'rx')
plt.xlabel('1st dimension')
plt.ylabel('2nd dimension')
```

```python
#Plotting the data "transformed" by PCA
class_0 = np.where(y==0)
class_1 = np.where(y==1)
plt.figure()
plt.title("Original data")
plt.plot(X_pca[class_0, 0], X_pca[class_0, 1], 'bo', mfc='none')
plt.plot(X_pca[class_1, 0], X_pca[class_1, 1], 'rx')
plt.title("Data transformed using PCA")
plt.xlabel('1st dimension')
plt.ylabel('2nd dimension')
```

```python
#Plotting the data transformed by KernelPCA
class_0 = np.where(y==0)
class_1 = np.where(y==1)
plt.figure()
plt.title("Original data")
plt.plot(X_kernel_pca[class_0, 0], X_kernel_pca[class_0, 1], 'bo', mfc='none')
plt.plot(X_kernel_pca[class_1, 0], X_kernel_pca[class_1, 1], 'rx')
plt.title("Data transformed using KernelPCA")
plt.xlabel('1st dimension')
plt.ylabel('2nd dimension')
```

```python
#Plotting the inverse of KernelPCA
class_0 = np.where(y==0)
class_1 = np.where(y==1)
plt.figure()
plt.title("Original data")
plt.plot(X_inverse[class_0, 0], X_inverse[class_0, 1], 'bo', mfc='none')
plt.plot(X_inverse[class_1, 0], X_inverse[class_1, 1], 'rx')
plt.title("Data inverse tranformed")
plt.xlabel('1st dimension')
plt.ylabel('2nd dimension')
```
